## Métodos

```python
import sympy as sy
import pandas as pd
import numpy as np
import plotly.express as px
from sympy.abc import x
```

```python
def bissec(a,b,f,error):
  tabela = pd.DataFrame(columns=['n','a','b','x','f(a)','f(b)','f(x)','|b-a|'])
  i=0
  while True:
    x= (a+b)/2 
    if f(a) * f(x) < 0:
        b = x
    else:
        a = x
    tabela.loc[i]=[i,a,b,x,f(a),f(b),f(x),abs(b-a)]
    i=i+1
    if ((abs(b-a)) < error) or (abs(f(x)) < error):
      break
  return x,i,tabela
```

```python
def derivate(f,var):
  x=sy.Symbol(var)
  f_l = f.diff(x)
  fi = x - (f/f_l)
  fi = sy.lambdify(x,fi,"numpy") 
  return fi
```

```python
def newton(x,f_s,f,var,error):
  fi=derivate(f_s,var)
  table = pd.DataFrame(columns=['n','x_n','f(x_n)','|x_n+1-x_n|'])
  i=0
  while True:
    fx=f(x)
    aux=fi(x)
    table.loc[i]=[i, x,f(x),abs(aux-x)]
    i+=1
    if abs(fx) < error  or abs(aux-x) < error:
        break
    x=aux
  return x,i,table
```

## Questão 1

Temos os pontos $x_0 = 0$, $y_0 = 5$, $x_f = 50$ e $y_f = 12$ e a equação $y=\frac{T_a}{w}cosh(\frac{w\cdot{x}}{T_a}) + y_0 -\frac{T_a}{w}$.

Substituindo as informações fornecidas podemos chegar na seguinte equação:

$$y=\frac{T_a}{w}cosh(\frac{w\cdot{x}}{T_a}) + y_0 -\frac{T_a}{w} <=>$$
$$\frac{T_a}{w}cosh(\frac{w\cdot{x}}{T_a}) + y_0 - y -\frac{T_a}{w} = 0 <=>$$
$$T_acosh(\frac{w\cdot{x}}{T_a}) + (y_0 - y)\cdot{w} -T_a = 0 <=>$$
$$T_acosh(\frac{10\cdot{x}}{T_a}) + (5 - y)\cdot{10} -T_a = 0 <=>$$
$$T_acosh(\frac{10\cdot{50}}{T_a}) + (5 - 12)\cdot{10} -T_a = 0 <=>$$
**$$T_acosh(\frac{500}{T_a}) -T_a -70= 0$$**

Para facilitar o uso das ferramentas chamaremos $T_a$ de x, então:
$$x\cdot{cosh(\frac{500}{x})} -x -70= 0$$

Consideraremos $f(x) = x\cdot{cosh(\frac{500}{x})} -x -70$, na qual devemos achar as raízes para encontrarmos a tensão horizontal necessária para equilibrar o cabo do exercício. Para isso, utilizaremos o método de newton para encontrarmos as raízes.



```python
#Vamos definir a função f(x):

fx = x*sy.cosh(500/x) -x -70;fx
```

```python
f=sy.lambdify(x,fx,'numpy')
```

```python
x=np.linspace(1500,2000,1000)
```

```python
fig = px.line (x = x,
               y = f(x),
               labels = {'y':'f(x)','x':'Valores de x'}, 
               template = 'presentation')
fig.show()
```

De acordo com o gráfico tiraremos valores de início para nosso método funcionar!

```python
erro = 0.01
x,i,table=newton(1770,fx,f,'x',erro)
print('Solução: {0:f} \nIterações: {1:d}'.format(x,i))
display(table)
```

Logo, a tensão horizontal necessária para equilibrar o peso do cabo é $T_a = 1797.261176$ 

## Questão 2

Na questão de número dois temos uma equação que simula a velocidade de ascenção de um foguete em um voo vertical:

$$v = u\cdot{ln(\frac{M_0}{M_0 - q\cdot{t}})} + g\cdot{t}$$

O problema pede para acharmos uma raíz contida no intervalo [6, 8] com tolerância de até 0,8% ($erro <= 0,008$).
Além disso, o problema nos fornece os seguintes dados:

*   $u = 200 m/s$
*   $M_0 = 1600 Kg$
*   $g = 9.8 m/s^2$
*   $q = 27 Kg/s$
*   $v = 100 m/s$

Logo, podemos achar a equação:

$$100 = 200\cdot{ln(\frac{1600}{1600 - 27\cdot{t}})} + 9.8\cdot{t}$$

$$200\cdot{ln(\frac{1600}{1600 - 27\cdot{t}})} + 9.8\cdot{t} - 100 = 0$$
-

Fazendo $t = x$, podemos dizer que $f(x) = 200\cdot{ln(\frac{1600}{1600 - 27\cdot{t}})} + 9.8\cdot{t} - 100$ e, dessa forma, acharemos a raíz dessa função no intervalo explicitado, então:

```python
#Vamos definir a função f(x):

fx = 200*sy.log(1600/(1600 - 27*x)) + 9.8*x - 100;fx
```

```python
f=sy.lambdify(x,fx,'numpy')
```

```python
#Fazendo os valores de x pertencerem ao intervalo pedido:
x=np.linspace(6,8,1000)
```

```python
fig = px.line (x = x,
               y = f(x),
               labels = {'y':'f(x)','x':'Valores de x'}, 
               template = 'presentation')
fig.show()
```

De acordo com o gráfico tiraremos valores de início para nosso método funcionar!

```python
a = 6
b = 8
erro = 0.008
x,i,tabela=bissec(a,b,f,erro)
print('Solução: {0:f} \nIterações: {1:d}'.format(x,i))
display(tabela)
```

Podemos ver que a raíz da equação que foi requisitada é $t = 7.460938$

## Questão 3

Na questão de número três um carro passa por um buraco e, o armotecedor faz o carro entrar em um movimento amortecido ao longo do tempo t, precisamos determinar os tempos para as 3 primeiras vezes que o carro passa pelo x(t) = 0, seguindo a equação e os dados fornecidos, temos:

$$x(t) = x_0\cdot{e^{-n\cdot{t}}}\cdot{(cos(p\cdot{t}) + \frac{n}{p}\cdot{sen(p\cdot{t}))}}$$

-

$$n = \frac{c}{2m}$$

-

$$p = \sqrt{\frac{k}{m} - \frac{c^2}{4m^2}}$$

-

Então:

$$x(t) = x_0\cdot{e^{-\frac{c}{2m}\cdot{t}}}\cdot{(cos(\frac{\sqrt{4mk - c^2}}{2m}\cdot{t}) + \frac{c}{\sqrt{4mk - c^2}}\cdot{sen(\frac{\sqrt{4mk - c^2}}{2m}\cdot{t}))}}$$


Se,



*   $k = 1.25\cdot{10^9} g/s^2$
*   $c = 1.4\cdot{10^7} g/s$
*   $m = 1.2\cdot{10^6} g$

Logo:

$$x(t) = x_0\cdot{e^{-5.8333\cdot{t}}}\cdot{(cos(31.7433\cdot{t}) + 0.1837\cdot{sen(31.7433\cdot{t}))}}$$

Como queremos determinar raízes de uma função $x(t)$, podemos trocar as dependências e reformular a função para que ela fique mais tranquila de se interpretar e calcular com funções do python, daí faremos:

$$f(x) = x_0\cdot{e^{-5.8333\cdot{t}}}\cdot{(cos(31.7433\cdot{x}) + 0.1837\cdot{sen(31.7433\cdot{x}))}}$$

Usando o método da bissecção podemos achar as 3 primeiras raízes (t cuja nossa referência é zerada $x(t) = 0$) pedidas pelo problema. Supondo que o buraco leve nosso ponto de referência chegar ao nível $x_0 = -0.2m$, logo:

$$f(x) = -0.2\cdot{e^{-5.8333\cdot{t}}}\cdot{(cos(31.7433\cdot{x}) + 0.1837\cdot{sen(31.7433\cdot{x}))}}$$

Com isso...



```python
#Vamos definir a função:

fx = -0.2*sy.exp(-5.8333*x)*(sy.cos(31.7433*x) + 0.1837*sy.sin(31.7433*x));fx
```

```python
f=sy.lambdify(x,fx,'numpy')
```

```python
x=np.linspace(0.001,0.3,1000)
```

```python
fig = px.line (x = x,
               y = f(x),
               labels = {'y':'f(x)','x':'Valores de x'}, 
               template = 'presentation')
fig.show()
```

De acordo com o gráfico tiraremos valores de início para nosso método funcionar!

```python
a = 0.001
b = 0.1
erro = 0.001
x,i,tabela=bissec(a,b,f,erro)
print('Solução: {0:f} \nIterações: {1:d}'.format(x,i))
display(tabela)
```

```python
a = 0.1
b = 0.2
erro = 0.001
x,i,tabela=bissec(a,b,f,erro)
print('Solução: {0:f} \nIterações: {1:d}'.format(x,i))
display(tabela)
```

```python
a = 0.2
b = 0.3
erro = 0.001
x,i,tabela=bissec(a,b,f,erro)
print('Solução: {0:f} \nIterações: {1:d}'.format(x,i))
display(tabela)
```

Encontramos os 3 primeiros tempos em que o ponto de referência volta ao ponto x(t) = 0, são eles:



*   $t_1 = 0.055141$

*   $t_3 = 0.253125$



## Questão 4

O nível de oxigênio de um rio é medido pela equação:

$$C = 10 - 15\cdot{(e^{-0.1x} - e^{-0.5x})}$$

Para saber qual a distância rio abaixo onde o nível de oxigênio é igual a 4, temos que igualar a equação a 4:

$$4 = 10 - 15\cdot{(e^{-0.1x} - e^{-0.5x})}$$

Com isso,

$$6 - 15\cdot{(e^{-0.1x} - e^{-0.5x})} = 0$$

Podemos fazer a equação igual a $f(x)$ e determinar sua raíz, que nos dará o ponto onde o nível de oxigênio é igual a 4, então:

$$f(x) = 6 - 15\cdot{(e^{-0.1x} - e^{-0.5x})}$$

Utilizando o método de newton, acharemos a distância x:

```python
#Definindo a função f(x):

fx = 6 - 15*(sy.exp(-0.1*x) - sy.exp(-0.5*x));fx
```

```python
f=sy.lambdify(x,fx,'numpy')
```

```python
x=np.linspace(0,10,1000)
```

```python
fig = px.line (x = x,
               y = f(x),
               labels = {'y':'f(x)','x':'Valores de x'}, 
               template = 'presentation')
fig.show()
```

Podemos ver que existem duas distâncias que rio abaixo podem ter nível de oxigênio igual a 4, então vamos subdividir o problema e aplicar duas vezes o método de newton.

```python
erro = 0.001
x,i,tabela=newton(0,fx,f,'x',erro)
print('Solução: {0:f} \nIterações: {1:d}'.format(x,i))
display(tabela)
```

```python
erro = 0.001
x,i,tabela=newton(7,fx,f,'x',erro)
print('Solução: {0:f} \nIterações: {1:d}'.format(x,i))
display(tabela)
```

Logo, podemos ver que os dois pontos rio abaixo que têm nível de oxigênio igual a 4 são:



*   $x_1 = 1.579960$
*   $x_2 = 8.870936$
